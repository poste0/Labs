### 2.7. создание нового поля вычисленного на основе значений других полей
#### 2.7.1. через выражение на базе имеющихся колонок,
#### Создадим столбец alco_ash, который будет равен среднему между alcohol и ash

```python
data['alco_ash'] = (data.alcohol + data.ash) / 2
data
```

#### 2.7.2. через DataFrame.apply
#### Сделаем такой же столбец, как и до этого, только через метод apply

```python
data['alco_ash'] = data.apply(lambda row: (row['alcohol'] + row['ash']) / 2, axis = 1)
data
```

#### 2.7.3. через Series.apply
#### Сделаем столбец на основе столбца alcohol

```python
data['alco_exp'] = data['alcohol'].apply(lambda elem: np.exp(elem))
data
```
