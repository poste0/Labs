```python
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
import sys
import sqlite3
```

```python
class Window(QMainWindow):
    def __init__(self):
        super().__init__()
        
        title = 'Laba 3'
        self.setWindowTitle(title)
        
        x = 1000
        y = 1000
        
        self.setGeometry(x, y, x, y)
        self.setFixedSize(x, y)
        
        menuBar = self.menuBar()
        menu = menuBar.addMenu("Menu")
        
        connect = QAction('Set connection', self)
        connect.triggered.connect(self.connectDb)
        
        close = QAction('Close connection', self)
        close.triggered.connect(self.closeDb)
        
        menu.addAction(connect)
        menu.addAction(close)
        
        buttonSizeX = 100
        buttonSizeY = 20
        
        self.bt1 = QPushButton(self)
        self.bt1.setText('bt1')
        self.bt1.move(20, 20)
        self.bt1.resize(buttonSizeX, buttonSizeY)
        self.bt1.clicked.connect(self.clickBt1)
        
        self.bt2 = QPushButton(self)
        self.bt2.setText('bt2')
        self.bt2.move(20, 40)
        self.bt2.resize(buttonSizeX, buttonSizeY)
        self.bt2.clicked.connect(self.clickBt2)
        
        self.bt3 = QPushButton(self)
        self.bt3.setText('bt3')
        self.bt3.move(200, 40)
        self.bt3.resize(buttonSizeX, buttonSizeY)
        self.bt3.clicked.connect(self.clickBt3)
        
        self.items = ['type', 'name', 'tbl_name', 'rootpage', 'sql']
        self.comboBox = QComboBox(self)
        self.comboBox.move(200, 20)
        self.comboBox.resize(buttonSizeX, buttonSizeY)
        self.comboBox.addItems(self.items)
        self.comboBox.currentTextChanged.connect(self.comboBoxClick)
        
        self.tabs = QTabWidget(self)
        self.tab1 = QTableWidget(self)
        self.tab2 = QTableWidget(self)
        self.tab3 = QTableWidget(self)
        self.tab4 = QTableWidget(self)
        self.tab5 = QTableWidget(self)
        
        self.tabs.addTab(self.tab1, 'Tab 1')
        self.tabs.addTab(self.tab2, 'Tab 2')
        self.tabs.addTab(self.tab3, 'Tab 3')
        self.tabs.addTab(self.tab4, 'Tab 4')
        self.tabs.addTab(self.tab5, 'Tab 5')
        
        self.tabs.move(20, 200)
        self.tabs.resize(1000, 500)
        
    def connectDb(self):
        self.db = sqlite3.connect('example.db')
        
        self.getAll(self.tab1)
    
    def addTableRow(self, table, data):
        row = table.rowCount()
        table.setRowCount(row + 1)
        
        col = 0
        for item in data:
            cell = QTableWidgetItem(str(item))
            table.setItem(row, col, cell)
            col += 1
    
    def clickBt1(self):
        if self.db != 0:
            self.remove(self.tab2)
            
            data = self.db.cursor().execute('SELECT type FROM sqlite_master').fetchall()
            
            self.tab2.setColumnCount(1)
            self.tab2.setHorizontalHeaderLabels(['type'])
            
            for row in data:
                self.addTableRow(self.tab2, row)
                
            self.tabs.setCurrentIndex(1)
            
    def getAll(self, tab):
        c = self.db.cursor()
        
        result = c.execute('SELECT * FROM sqlite_master').fetchall()
        result_info = c.execute('PRAGMA table_info(sqlite_master)').fetchall()
        
        tab.setColumnCount(len(result_info))
        tab.setHorizontalHeaderLabels([row[1] for row in result_info])
        
        for row in result:
            self.addTableRow(tab, row) 
    
    def clickBt2(self):
        self.remove(self.tab4)
        self.getAll(self.tab4)
        
        self.tabs.setCurrentIndex(3)
    
    def clickBt3(self):
        self.remove(self.tab5)
        self.getAll(self.tab5)
        
        self.tabs.setCurrentIndex(4)
    
    def remove(self, tab):
        while(tab.rowCount() > 0):
            tab.removeRow(0)
    
    def closeDb(self):
        self.remove(self.tab1)
        self.remove(self.tab2)
        self.remove(self.tab3)
        self.remove(self.tab4)
        self.remove(self.tab5)
        
        self.db.close()
        
        self.db = 0
    
    def comboBoxClick(self, value):
        if self.db != 0:
            self.remove(self.tab3)
            
            data = self.db.cursor().execute('SELECT ' + value + ' FROM sqlite_master').fetchall()
            
            self.tab3.setColumnCount(1)
            self.tab3.setHorizontalHeaderLabels([value])
            
            for row in data:
                self.addTableRow(self.tab3, row)
                
            self.tabs.setCurrentIndex(2)
```

```python
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Window()
    window.show()
    app.exec()
```
